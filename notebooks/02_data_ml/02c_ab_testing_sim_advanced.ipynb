Simulate A/B outcomes, compute lifts, t-tests, and bootstrap CIs. Discuss guardrails.



```python
import numpy as np
from utils.ab_stats import two_sample_ttest, bootstrap_ci
from utils.grading import run_checks, assert_true

rng = np.random.default_rng(0)
A = rng.normal(loc=0.0, scale=1.0, size=500)
B = rng.normal(loc=0.2, scale=1.0, size=500)

lift = float(B.mean() - A.mean())
t, p = two_sample_ttest(A, B)
ci = bootstrap_ci(B - A, np.mean, n_boot=1000, alpha=0.05, rng=0)

chk1 = lambda: assert_true("detect lift", lift > 0)
chk2 = lambda: assert_true("p<0.05 typical", p < 0.05)
chk3 = lambda: assert_true("ci covers lift", ci[0] <= lift <= ci[1])
run_checks(chk1, chk2, chk3)

{"lift": lift, "t": t, "p": p, "ci": ci}

```
