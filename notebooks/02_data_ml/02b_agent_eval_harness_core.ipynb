# 02b — Agent Evaluation Harness (core)

Timebox: 45–60 minutes

Evaluate stubbed agent behaviors against golden expectations and detect edge cases.



```python
import json
from pathlib import Path
from typing import Dict, Any

from utils.datasets import data_path
from utils.grading import run_checks, assert_true, assert_equal

# Stub agent with deterministic rules

def agent_respond(prompt: str, tools: Dict[str, Any]) -> str:
    p = prompt.lower()
    if "2 and two" in p or "2+2" in p:
        return "4"
    if "atlantis" in p:
        return "Unknown / mythical"
    if "weather" in p:
        return "Graceful timeout handling"
    return "I don't know"

cases_path = data_path("agent_cases/agent_cases.jsonl")

results = []
with open(cases_path, "r", encoding="utf-8") as f:
    for line in f:
        c = json.loads(line)
        out = agent_respond(c["prompt"], {t: None for t in c.get("tools", [])})
        results.append({"case_id": c["case_id"], "out": out, "golden": c["golden"]})

# Checks
ok = sum(1 for r in results if r["out"] == r["golden"])
chk1 = lambda: assert_true("all cases matched", ok == len(results))
run_checks(chk1)
results
```
