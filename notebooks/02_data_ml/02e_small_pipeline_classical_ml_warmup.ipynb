# 02e — Small ETL + Classical ML Pipeline (warmup)

Timebox: 30–45 minutes

Build a tiny pipeline: load CSV, split, scale, train logistic regression, evaluate.



```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

from utils.datasets import read_csv
from utils.grading import run_checks, assert_true

df = read_csv("toy/toy_classification.csv")
X = df[["f1","f2","f3"]].values
y = df["label"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(max_iter=1000))
])
pipe.fit(X_train, y_train)

pred = pipe.predict(X_test)
acc = accuracy_score(y_test, pred)

chk1 = lambda: assert_true("reasonable acc", acc >= 0.5)
run_checks(chk1)

acc
```
